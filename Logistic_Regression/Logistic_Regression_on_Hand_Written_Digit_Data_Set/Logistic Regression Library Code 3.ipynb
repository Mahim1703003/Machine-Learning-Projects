# Logistic Regression on hand written digit dataset (with standardization)

## Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset

```python
from sklearn.datasets import load_digits
```

```python
digit = load_digits()
```

```python
digit
```

```python
df = pd.DataFrame(data = digit['data'],columns=digit['feature_names'])
df
```

```python
df['target'] = digit['target']
```

```python
df
```

```python
df[df.columns[:-1]]
```

```python
x = df[df.columns[:-1]]
x
```

```python
y = df[df.columns[-1:]]
y
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x = scaler.transform(x)
```

```python
x = pd.DataFrame(x, columns=df.columns[:-1])
x
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.23,random_state=2)
```

```python
x_train
```

```python
y_train
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='saga',multi_class='ovr',max_iter=500)
```

```python
model.fit(x_train, y_train)
```

## Test Accuracy

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
