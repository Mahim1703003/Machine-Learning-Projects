# Linear Regression

```python
np.array(x_train)
```

```python
X = np.array(x_train)
X
```

```python
np.transpose(X)
```

```python
Xt = np.transpose(X)
Xt
```

```python
XtX = np.dot(Xt,X)
XtX
```

```python
XtXinverse = np.linalg.inv(XtX)
XtXinverse
```

```python
b = np.dot(XtXinverse,Xt)
b
```

```python
np.array(y_train)
```

```python
Y = np.array(y_train)
Y
```

```python
b = np.dot(b,Y)
b
```

```python
b.shape
```

```python
b.ndim
```

```python
n = len(x_train.columns)
n
```

```python
print('intercept :',b[0,0])
```

```python
print('coefficients :')
for i in range(1,n):
    print('b'+str(i),':',b[i,0])
```

```python
b
```

```python
x_test
```

```python
y_test
```

```python
# y_prediction

y_cap = list()
y_prediction = 0

for row in x_test.index:
    for i in range(len(x_test.columns)):
        y_prediction = y_prediction + b[i,0] * x_test[x_test.columns[i]][row]
    
    y_cap.append(y_prediction)
    y_prediction = 0

y_cap
```

```python
len(y_cap)
```

## Model Evaluation on Test Dataset

```python
y_test
```

```python
# R Square

y_mean = y_test[y_test.columns[0]].mean()

sum1 = 0
sum2 = 0

for i in range(len(y_cap)):
    diff = y_test.iloc[i][y_test.columns[0]] - y_cap[i] 
    sum1 = sum1 + (diff ** 2)

for i in y_test.index:
    diff = y_test[y_test.columns[0]][i] - y_mean
    sum2 = sum2 + (diff ** 2)

r2 = 1 - (sum1 / sum2)

print(r2)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_cap) # first comes actual value and then predicted value
print(r2)
```

```python
# MSE

sum1 = 0

for i in range(len(y_cap)):
    diff = y_test.iloc[i][y_test.columns[0]] - y_cap[i] 
    sum1 = sum1 + (diff ** 2)

mse = sum1 / (len(y_cap))
print(mse)
```

```python
# RMSE

rmse = np.sqrt(mse)
rmse
print(rmse)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_cap)
print('MSE :',mse)
rmse = np.sqrt(mse)
print('RMSE :',rmse)
```
